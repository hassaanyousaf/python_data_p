```python
import pandas as pd
from datasets import load_dataset
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset["train"].to_pandas()
df['job_posted_date']=pd.to_datetime(df.job_posted_date)
```

```python
#median salary analysis by country
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

```python
#top six countries with the highest median salaraies
import matplotlib.pyplot as plt

top_countries=df['job_country'].value_counts().head(6).index

top_country_salary=df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')

top_country_salary=top_country_salary.loc[top_countries]
jobs=['Data Analyst', 'Data Scientist','Data Engineer']
top_country_salary=top_country_salary[jobs]
top_country_salary.plot(kind='bar')

plt.title('Median salary by country and job title (USD)')
plt.ylabel('Median salary by country and job title (USD)')
plt.xlabel('')
```

```python
# Number of job postings each job title per month
df['job_posted_month'] = df['job_posted_date'].dt.month
pivot_postings = df.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc='size'
)
jobs=['Data Analyst', 'Data Scientist','Data Engineer']
pivot_postings=pivot_postings[jobs]
pivot_postings.plot(kind='bar')
```
